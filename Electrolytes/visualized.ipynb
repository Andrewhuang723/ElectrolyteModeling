## Convert .MOL to .lt file

```python
system_block = """
write_once("Data Boundary") {
  -10 10 xlo xhi
  -10 10 ylo yhi
  -10 10 zlo zhi
}

# Create an input deck for LAMMPS.
write_once("In Init"){
 # Input variables.
 variable run    string ec   # output name
 variable ts     equal  1         # timestep
 variable temp   equal  300       # equilibrium temperature
 variable p      equal  1.        # equilibrium pressure
 variable d      equal  1000      # output frequency
 variable equi   equal  5000      # Equilibration steps
 variable prod   equal  30000     # Production steps

 # PBC (set them before the creation of the box).
 boundary p p p
}

# Run an NPT simulation.
write_once("In Run"){
 # Derived variables.
 variable tcouple equal \$\{ts\}*100
 variable pcouple equal \$\{ts\}*1000

 # Output.
 thermo          \$d
 thermo_style custom step etotal evdwl ecoul elong ebond eangle &
 edihed eimp ke pe temp press vol density cpu
 thermo_modify flush yes

 # Trajectory.
 dump TRJ all dcd \$d \$\{run\}.dcd
 dump_modify TRJ unwrap yes

 # Thermalisation and relaxation, NPT ensemble.
 timestep       \$\{ts\}
 fix             NPT all npt temp \$\{temp\} \$\{temp\} \$\{tcouple\} iso \$p \$p \$\{pcouple\}
 velocity all create \$\{temp\} 858096 dist gaussian
 # Short runs to update the PPPM settings as the box shinks.
 run    \$\{equi\} post no
 run    \$\{equi\} post no
 run    \$\{equi\} post no
 run    \$\{equi\}
 # From now on, the density shouldn't change too much.
 run    \$\{prod\}
 unfix NPT
}
"""

def MOL2LT(molecule_name, mol, generate_system_file=True):
    positions = mol.GetConformer().GetPositions()

    with open(f"./{molecule_name}/{molecule_name}.lt", "w") as f:

        header = f"# {molecule_name}.lt\n\n"
        f.write(header)

        f.write("import /Users/andrewhuang/Desktop/moltemplate/moltemplate/force_fields/oplsaa.lt\n")

        f.write(f"{molecule_name} inherits OPLSAA {{\n")
        f.write("\n")
        f.write(f"  write(\"Data Atoms\") {{\n")

        for atom in mol.GetAtoms():
            atom_postion = f"{positions[atom.GetIdx()][0]} {positions[atom.GetIdx()][1]} {positions[atom.GetIdx()][2]}"
            f.write(f"      $atom:{atom.GetIdx()+1}    $mol:.  @atom:      0.0     {atom_postion}\n")
        f.write(f'  }}\n')
        f.write('\n')


        f.write(f"  write(\"Data Bond List\") {{\n")
        for bond in mol.GetBonds():
            f.write(f"      $bond:{bond.GetIdx()+1}  $atom:{bond.GetBeginAtomIdx()+1}   $atom:{bond.GetEndAtomIdx()+1}\n")
        f.write(f'  }}\n')
        f.write('\n')

        f.write('}')
    
    print(f"Molcule {molecule_name} was converted to {molecule_name}.lt successfully.")


    if generate_system_file:
        with open(f"./{molecule_name}/system.lt", "w") as f:
            f.write(f"import \"{molecule_name}.lt\"\n")
            f.write("\n")
            f.write(f"{molecule_name} = new {molecule_name}[1].move(5.0, 0, 0)\n")
            f.write("       [1].move(0, 5.0, 0)\n")
            f.write("       [1].move(0, 0, 5.0)\n")
            f.write("\n")
            f.write(system_block)

    print(f"{molecule_name}/system.lt generated.")
```

## Convert data (.data) to molecule file (.txt)

```python

# molecule = "DMC"

def DATA2MOLTXT(molecule):

    Headers = [f"# {molecule} molecule\n"]
    Coords = []
    Types = []
    Charges = []
    Bonds = []
    Angles = []
    Dihedrals = []
    Impropers = []
    flag = None
    
    with open(f"./{molecule}/system_after.data", "r") as f:
        for i, line in enumerate(f.readlines()):      
            elements = line.strip()
            str_list = line.rsplit()
            if elements == "":
                continue
            if len(str_list) > 0 and (str_list[-1] in ["atoms", "bonds", "angles", "dihedrals", "impropers"]):
                Headers.append(elements+'\n')
            if flag == "Atom" and len(str_list) > 1:
                Coords.append(f"{str_list[0]} {str_list[4]} {str_list[5]} {str_list[6]}\n")
                Types.append(f"{str_list[0]} {str_list[2]}\n")
                Charges.append(f"{str_list[0]} {str_list[3]}\n")
            elif flag == "Bonds" and len(str_list) > 1:
                Bonds.append(elements + '\n')
            elif flag == "Angles" and len(str_list) > 1:
                Angles.append(elements + '\n')
            elif flag == "Dihedrals" and len(str_list) > 1:
                Dihedrals.append(elements + '\n')
            elif flag == "Impropers" and len(str_list) > 1:
                Impropers.append(elements + '\n')
            else:
                pass
    
            if elements == "Atoms # full":
                flag = "Atom"
                Coords.append('\nCoords\n\n')
                Charges.append('\nCharges\n\n')
                Types.append('\nTypes\n\n')
    
            elif elements == "Velocities":
                flag = None
            elif elements == "Bonds":
                flag = "Bonds"        
                Bonds.append('\nBonds\n\n')
            elif elements == "Angles":
                flag = "Angles"
                Angles.append('\nAngles\n\n')
            elif elements == "Dihedrals":
                flag = "Dihedrals"
                Dihedrals.append('\nDihedrals\n\n')
            elif elements == "Impropers":
                flag = "Impropers"
                Impropers.append("\nImpropers\n\n")
            else:
                pass
        
    
    
    
    with open(f"./{molecule}/molecule.txt", "w") as target:
        for block in [Headers, Coords, Charges, Types, Bonds, Angles, Dihedrals, Impropers]:
            for line in block:
                target.write(line)
    
    print(f"Data file of {molecule} has converted to molecule.txt")
```
