### Convert data to molecule file

```python
# molecule = "PMMA"
 
Headers = [f"# EC DEC molecule\n"]
Coords = []
Types = []
Charges = []
Bonds = []
Angles = []
Dihedrals = []
Impropers = []
flag = None
 
with open(f"./Mixing/minimization/system.data", "r") as f:
    for i, line in enumerate(f.readlines()):      
        elements = line.strip()
        str_list = line.rsplit()
        if elements == "":
            continue
        if len(str_list) > 0 and (str_list[-1] in ["atoms", "bonds", "angles", "dihedrals", "impropers"]):
            Headers.append(elements+'\n')
        if flag == "Atom" and len(str_list) > 1:
            Coords.append(f"{str_list[0]} {str_list[4]} {str_list[5]} {str_list[6]}\n")
            Types.append(f"{str_list[0]} {str_list[2]}\n")
            Charges.append(f"{str_list[0]} {str_list[3]}\n")
        elif flag == "Bonds" and len(str_list) > 1:
            Bonds.append(elements + '\n')
        elif flag == "Angles" and len(str_list) > 1:
            Angles.append(elements + '\n')
        elif flag == "Dihedrals" and len(str_list) > 1:
            Dihedrals.append(elements + '\n')
        elif flag == "Impropers" and len(str_list) > 1:
            Impropers.append(elements + '\n')
        else:
            pass
 
        if elements == "Atoms # full":
            flag = "Atom"
            Coords.append('\nCoords\n\n')
            Charges.append('\nCharges\n\n')
            Types.append('\nTypes\n\n')
 
        elif elements == "Velocities":
            flag = None
        elif elements == "Bonds":
            flag = "Bonds"        
            Bonds.append('\nBonds\n\n')
        elif elements == "Angles":
            flag = "Angles"
            Angles.append('\nAngles\n\n')
        elif elements == "Dihedrals":
            flag = "Dihedrals"
            Dihedrals.append('\nDihedrals\n\n')
        elif elements == "Impropers":
            flag = "Impropers"
            Impropers.append("\nImpropers\n\n")
        else:
            pass
       
 
 
 
with open(f"./MixingPMMA/solvent.txt", "w") as target:
    for block in [Headers, Coords, Charges, Types, Bonds, Angles, Dihedrals, Impropers]:
        for line in block:
            target.write(line)
```

## Mixing EC & DEC

### Calculate entropy

```python
Step = []
Entropy = []
with open("./Mixing/equilibration/dump.lammpstrj", "r") as f:
    flag = None
    batch = []
    cnt = 0
    for i, line in enumerate(f.readlines()):
        if line == "ITEM: TIMESTEP\n":
            flag = "step"
            if len(batch) > 0:
                Entropy.append(batch) 
                batch = []
            continue

        if line == "ITEM: ATOMS x y z type id c_my_entropy\n":
            flag = "entropy"
            continue

        if flag == "step":
            Step.append(int(line.rsplit()[0]))
            flag = None
        
        if flag == "entropy":
            s = line.rsplit()[-1]
            if s != 'nan':
                batch.append(float(s))

Entropy.append(batch)
Entropy = list(map(lambda x: sum(x) * -1, Entropy))

# pmma_entropy = 34.98 * 1
dec_entropy = 119.2223 * 100
ec_entropy = 108.4058 * 100

delta_S_mix = list(map(lambda x: (x / (4.2) - (dec_entropy+ec_entropy))/1000, Entropy)) # cal/mol K
```

### Enthalpy

```python
import numpy as np

Enthalpy = []
with open("./Mixing/equilibration/log.lammps", "r") as f:
    for i, line in enumerate(f.readlines()):
        if i >= 4144 and i <= 4204:
            enthalpy = float(line.rsplit()[2])
            Enthalpy.append(enthalpy)



# pmma_enthalpy = 233.7869 * 1
dec_enthalpy = 12.789 * 100
ec_enthalpy = 7.1734 * 100

delta_H_mix = list(map(lambda x: x - (dec_enthalpy+ec_enthalpy), Enthalpy))
delta_H_mix = np.array(delta_H_mix[1:]).reshape(-1, 2).mean(axis=1)
```

### Gibbs free energy

```python
delta_G_mix = np.array(delta_H_mix) + 300 * np.array(delta_S_mix[1:])
```
